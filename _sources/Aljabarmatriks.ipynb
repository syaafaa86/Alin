C. Operasi Perkalian matriks adalah operasi mengalikan  matriks mengalikan setiap elemen baris pada matriks pertama dengan elemen kolom pada matriks kedua. syarat mengalikan matriks itu kolom dari matriks pertama harus sama dengan baris dari matriks kedua misal matriks A ordo nya 2x3 dan matriks B ordo 3x3

```python
# contoh: (AXD)
A = np.array([[3,2,4],[1,3,2],[2,3,5]])
print(A)
print()
D = np.array([[1,1,4],[1,3,2],[2,1,2]])
print(D)
```

```python
result = [[0, 0, 0],
          [0, 0, 0],
          [0, 0, 0]]

# Perkalian matriks secara manual
for i in range(len(A)):  # Loop baris matriks A
    for j in range(len(D[0])):  # Loop kolom matriks B
        for k in range(len(D)):  # Loop kolom matriks A/baris matriks B
            result[i][j] += A[i][k] * D[k][j]

# Menampilkan hasil
print("Hasil perkalian matriks A dan D:")
for row in result:
    print(row)
```

3. Mengalikan matriks A dengan matriks B

```python
A = np.array([[3,2,4],[1,3,2],[2,3,5]])
print(A)
print()
B = np.array([[1,1,2],[1,4,2],[1,1,3]])
print(B)
```

```python
result = [[0, 0, 0],
          [0, 0, 0],
          [0, 0, 0]]

# Perkalian matriks secara manual
for i in range(len(A)):  # Loop baris matriks A
    for j in range(len(B[0])):  # Loop kolom matriks B
        for k in range(len(B)):  # Loop kolom matriks A/baris matriks B
            result[i][j] += A[i][k] * B[k][j]

# Menampilkan hasil
print("Hasil perkalian matriks A dan B:")
for row in result:
    print(row)
```
